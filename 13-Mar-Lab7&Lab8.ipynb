```python
# Qn 3

male_professors = sal[sal['sex'] == 'Male']
female_professors = sal[sal['sex'] == 'Female']

highest_male_salary = male_professors.loc[male_professors['salary'].idxmax()]
lowest_male_salary = male_professors.loc[male_professors['salary'].idxmin()]

highest_female_salary = female_professors.loc[female_professors['salary'].idxmax()]
lowest_female_salary = female_professors.loc[female_professors['salary'].idxmin()]
```

```python
# Qn 4

highest_salary_index = sal['salary'].idxmax()
lowest_salary_index = sal['salary'].idxmin()

professor_highest_salary = sal.loc[highest_salary_index]
professor_lowest_salary = sal.loc[lowest_salary_index]
```

```python
print("Highest: ")
print(professor_highest_salary)
print()
print("Lowest: ")
print(professor_lowest_salary)
```

```python
# Qn 7

import matplotlib.pyplot as plt

gender_counts = df['sex'].value_counts()

print("No. of male staff: ", gender_counts['Male']);
print("No. of female staff: ", gender_counts['Female']);

plt.figure(figsize=(8,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct="%1.1f%%", startangle=20)
plt.title("Male and Female staffs plotted in pie chart\n")
plt.axis("equal")
plt.show()
```

```python
# Qn 8

rank_counts = df['rank'].value_counts()

print("No. of Profs: ", rank_counts['Prof']);
print("No. of Asst. Prof: ", rank_counts['AsstProf']);
print("No. of Associate. Prof: ", rank_counts['AssocProf']);

plt.figure(figsize=(8,6))
plt.pie(rank_counts, labels=rank_counts.index, autopct="%1.1f%%", startangle=20)
plt.title("Profs in pie chart\n")
plt.axis("equal")
plt.show()
```

```python
# Qn 9

df_sorted = df.sort_values(by='service')

# Find the junior-most employee (first row after sorting)
junior_employee = df_sorted.iloc[0]

# Find the senior-most employee (last row after sorting)
senior_employee = df_sorted.iloc[-1]

# Print the information of the junior-most and senior-most employees
print("Junior-most Employee:")
print(junior_employee)
print("\nSenior-most Employee:")
print(senior_employee)
```

```python
# Qn 10

bin_edges = list(range(50000, int(df['salary'].max()) + 15000, 15000))

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['salary'], bins=bin_edges, edgecolor='white')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Histogram of Salaries')
plt.xticks(bin_edges, rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
